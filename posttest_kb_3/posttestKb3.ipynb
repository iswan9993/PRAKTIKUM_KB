1. disini saya mengimport beberapa libraries untuk membaca file.csv dan untuk melakukan visualisasi data seperti matplotlib.dimana pada file csv saya memiliki variable dengan nama data_weather untuk membaca file 'IndianWeatherRepository.csv'.sehingga muncul seperti pada output yang terlihat 

```python
import matplotlib.pyplot as plt
import pandas as pd

data_weather = pd.read_csv('IndianWeatherRepository.csv')


data_weather.head()

```

2. untuk melakukan visualisasi data kita menggunakan fungsi groupby agar mempermudah kita dalam melakukan visualisasi data yang kita inginkan seperti contoh saya melalukan groub untuk data 'location_name dan temperature_celcius'. dimana kita juga menggunakan fungsi mean() untuk melakukan perhitungan rata-rata temperature untuk setiap lokasi yang ditentukan dalam perhitungan celcius.karena kita akan membatasi agar data yang tampil itu tidak terlalu banyak maka kita akan menampilkan empat data pertama dengan menggunakan fungsi 'head(4)'.

```python
data_temperature = data_weather.groupby('location_name')['temperature_celsius'].mean()
data_temperature = data_temperature.head(4)
```

3. kita memberikan labelx dan labely pada setiap atribut dimana 'data_temperature.index'mengacu pada label yang dalam hal ini adalah location_name dan 'data_temperature.values' yaitu data yang akan berada pada garis Y dalam diagram yang dalam hal ini temperatur yang mengembalikan nilai rata- pada nama lokasinya.

```python
x = data_temperature.index
y = data_temperature.values
```

 
### 1. Diagram Batang

4. selanjutnya kita melakukan menggunakan fungsi matplotlib untuk menggambar diagram batang dimana:
    plt.bar(x, y, color=['green', 'blue', 'yellow', 'red'])==> untuk membuat diagram batangnya dimana parameter x membentuk sumbu yang dalam hal ini adalah nama lokasinya dan sumbu Y adalah temperatur suhu rata-rata setiap lokasinya serta memberikan warna pada setiap diagramnya. plt.title('Rata-rata Temperatur per Lokasi')====> digunakan untuk membuat judul pada grafik.plt.xlabel('Lokasi')===> digunakan untuk membuat title pada lokasinya atau pada sumbu X.  plt.ylabel('Rata-rata Temperatur (°C)')====> digunakan untuk membuat tittle pada label Y atau sumbu Y. plt.xticks(rotation=45) ====> digunakan untuk membuat nama lokasinya miring 45 derajat. plt.show()===> untuk menampilkan diagram. bisa dilihat outputnya.

```python
plt.bar(x, y, color=['green', 'blue', 'yellow', 'red'])


plt.title('Rata-rata Temperatur per Lokasi')
plt.xlabel('Lokasi')
plt.ylabel('Rata-rata Temperatur (°C)')


plt.xticks(rotation=45) 
plt.show()
```

### Histogram

#### kesimpulan

1. pada histogram kita dapat melihat bahwa seberapa sering sebuah data itu muncul. kita dapat mebuat sebuah istogram dengan cara plt.hist(temperature_data, bins=15, color='green') dimana plt.histuntuk membuat sebuah histogram diikuti dengan nama atribut yang ingin kita masukkan kedalam histogram, menggunakan bins untuk bnyak rentang data yang akan kita gunakan serta color untuk pemilihan warna pada histogram kita.  sama seperti sebelumya dimana plt.title digunakan untuk membuat sebuah judul, xlabel untuk membuat keterangan pada sumbu X dan ylabel digunakan untuk membuat keterangan pada sumbu Y. bisa dilihat dari datanya dimana temperatur dianta 20 derajat dan 30 derajat lebih sering muncul daripada yang lain.

```python
temperature_data = data_weather['temperature_celsius']

plt.hist(temperature_data, bins=15, color='green')

plt.title('Histogram Temperatur')
plt.ylabel('Frekuensi kemunculan')
plt.xlabel('Temperatur (°C)')
plt.show()
```

### Pie Chart

#### Kesimpulan 

pada pie chart kita bisa melihat dengan jelas kumpulan data kit dalam bentuk lingkaran. untuk membuatnya kita pertama memilih atribut apa yang akan kita gunakan dalam pie chart tersebut. menggunakan fungsi data_temperature = data_weather['location_name'].value_counts().head(4) yaitu menghitung jumlah data perlokasi dan hasilnya banyaknya data dalam satu lokasi yang sama dengan value_counts().sedangkan fungsi head(4) yaitu mengambil 4 data teratas dalam atribut tersebut. untuk membuat visualisasi data pie chart kita menggunakan fungsi 'plt.pie' ,labels=data_temperature.index adalah keterangan untuk nama lokasi, 'autopct="%1.1f%%' adalah format persentase yang ditampilkan didalam sebuah pie chart artinya satu angka desimal sebelum titik dan satu angka desimal setelah titik, sedangkan plt.title digunakan untuk memberikan keterangan pada judul pie chart kita

```python
data_temperature = data_weather['location_name'].value_counts().head(4)
plt.pie(data_temperature, labels=data_temperature.index, autopct="%1.1f%%")
plt.title('Jumlah Data per Lokasi')

plt.show()
```

### Heatmap

fungsi grouby dibawah adalah untuk melihat kumpulan data dalam location name apa saja yang akan kita masukkan kedalam heatmap nanti.

```python
data_temperature = data_weather.groupby('location_name').mean()
data_temperature
```

#### kesimpulan

Heatmap adalah representasi visual dari data tabular di mana nilai-nilai dalam tabel diwakili oleh warna. Biasanya, heatmap digunakan untuk memvisualisasikan matriks atau tabel dengan menggunakan warna untuk menunjukkan intensitas atau nilai dari masing-masing sel dalam matriks.Pada heatmap, setiap sel dalam matriks direpresentasikan oleh warna yang mencerminkan nilai dari sel tersebut. Skala warna yang digunakan biasanya diatur sehingga warna-warna yang berbeda mencerminkan rentang nilai yang berbeda, memungkinkan pengamat untuk dengan cepat melihat pola atau distribusi dalam data. untuk menampilkan heatmap kita perlu mengimport library yang deperlukan yaitu seaborn trlebih dahulu. 'pilih_variables' ini digunakan untuk memilih atribut yang akan kita masukkan datanya kedalam heatmap, fungsi plt.figure untuk mengatur gambar yang ada di dalam heatmap, sns.heatmap(): Membuat heatmap. Parameter pertama adalah matriks korelasi yang ingin ditampilkan. Parameter cmap='coolwarm' digunakan untuk memilih skema warna (coolwarm digunakan di sini). Parameter annot=True akan menampilkan angka korelasi di dalam sel. Parameter center=0 digunakan untuk mengatur warna tengah heatmap ke 0. plt.title digunakan untuk membuat judul yang ada dalam heatmap.

```python
import seaborn as sns

pilih_variables = ['temperature_celsius', 'temperature_fahrenheit', 'wind_mph', 'wind_kph', 'wind_degree','pressure_mb','pressure_in','gust_kph','air_quality_Carbon_Monoxide']
cor_matrix_subset = data_weather[pilih_variables].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(cor_matrix_subset, cmap='coolwarm', annot=True, center=0)

plt.title('Heatmap Korelasi')
plt.show()
```
