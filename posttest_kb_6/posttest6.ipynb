langkah selanjunya kita akan membangun model dimana Dalam bagian "Input Layer," sebuah lapisan Dense (fully connected layer) dengan 16 neuron digunakan sebagai lapisan input. Aktivasi "relu" (Rectified Linear Unit) diterapkan pada setiap neuron dalam lapisan ini, dan jumlah input adalah 39, yang mengindikasikan bahwa data input memiliki 39 fitur.Kemudian, ada beberapa "Hidden Layers" dengan berbagai jumlah neuron (misalnya, 32, 64, 128) dan aktivasi "relu." Ini adalah lapisan tersembunyi yang membantu model mempelajari pola-pola yang lebih kompleks dalam data. Adanya beberapa lapisan tersembunyi ini dapat meningkatkan kapasitas model, tetapi juga meningkatkan kompleksitas dan potensi overfitting.Terakhir, ada "Output Layer" dengan satu neuron dan aktivasi "sigmoid." Lapisan ini digunakan dalam tugas klasifikasi biner, dan aktivasi sigmoid menghasilkan probabilitas kelas positif. Jumlah neuron dalam output layer harus disesuaikan dengan jumlah kelas atau kategori yang ingin diprediksi.Hasil dari model.summary() adalah ringkasan yang mencantumkan detail arsitektur jaringan saraf tiruan, termasuk jumlah neuron di setiap lapisan, jumlah parameter yang dapat di-train, dan penggunaan aktivasi.

```python
from keras import Sequential
from keras.layers import Dense

model = Sequential()

#Input Layer
model.add(Dense(16, activation='relu', input_dim=39))

#Hidden Layer
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))






#Output Layer
model.add(Dense(1, activation='sigmoid')) #output layer harus disamakan dengan jumlah class, semakin banyak jumlah neuron semakin kompleks modelnya jadi bisa overfitting

model.summary()
```

selanjutnya kita mengimport beberapa library dimana from keras.utils import to_categorical: Fungsi to_categorical digunakan untuk mengkodekan target kategori menjadi bentuk yang sesuai dengan tugas klasifikasi. Fungsi ini akan mengubah label kategori menjadi representasi one-hot encoding.
from keras.optimizers import Adam: Ini mengimpor optimizer Adam, yang adalah salah satu optimizer yang populer digunakan dalam pelatihan jaringan saraf tiruan. Optimizer adalah algoritma yang digunakan untuk mengoptimalkan bobot dan bias jaringan selama proses pelatihan. Adam adalah metode optimasi yang efisien dan sering digunakan dalam berbagai tugas.from keras.losses import binary_crossentropy: Ini mengimpor fungsi kerugian (loss function) binary_crossentropy, yang umumnya digunakan dalam tugas klasifikasi biner. Kerugian ini digunakan untuk mengukur kesalahan antara prediksi model dan nilai target. Tujuannya adalah untuk meminimalkan kerugian ini selama pelatihan model agar model dapat belajar dengan baik.

```python
from keras.utils import to_categorical
from keras.optimizers import Adam
from keras.losses import binary_crossentropy
```

EarlyStopping dari Keras, yang digunakan untuk mengimplementasikan teknik penghentian dini (early stopping) selama pelatihan model jaringan saraf tiruan.monitor='accuracy': Ini mengindikasikan bahwa kita akan memonitor metrik akurasi (accuracy) selama pelatihan model. patience=20: Patience adalah jumlah epoch yang akan diizinkan tanpa peningkatan dalam akurasi sebelum pelatihan dihentikan.

```python
from keras.callbacks import EarlyStopping

monitor_val_acc = EarlyStopping(monitor='accuracy', 
                                patience=20)
```

optimizer='Adam'Optimizer ini bertanggung jawab untuk mengoptimalkan bobot dan bias dalam model selama pelatihan dengan tujuan meminimalkan fungsi kerugian.loss=tf.keras.losses.BinaryCrossentropy(): Ini adalah fungsi kerugian yang akan digunakan selama pelatihan. BinaryCrossentropy, yang merupakan kerugian umum yang digunakan dalam tugas klasifikasi biner. Kerugian ini mengukur sejauh mana prediksi model cocok dengan label target. metrics=['accuracy']untuk memantau sejauh mana model dapat mengklasifikasikan data dengan benar.

```python
model.compile(
    optimizer='Adam',
    loss = tf.keras.losses.BinaryCrossentropy(),
    metrics = ['accuracy']
)
```

x = x_scaled_train: Ini adalah data input (fitur) yang akan digunakan selama pelatihan. Data input harus sudah diproses dan diubah menjadi skala yang sesuai. y = y_train: Ini adalah label atau target yang sesuai dengan data input x_scaled_train. Model akan belajar untuk memprediksi label-label ini selama pelatihan. validation_data=(x_scaled_test, y_test): Ini adalah data uji yang digunakan untuk memantau kinerja model selama pelatihan.epochs=500: Ini adalah jumlah iterasi (epoch) yang akan digunakan selama pelatihan. Model akan melalui seluruh data pelatihan sebanyak 500 kali. verbose=1: Ini mengontrol tingkat keluaran selama pelatihan. Nilai 1 menghasilkan keluaran yang lengkap dan informatif, sementara nilai 0 akan membuat keluaran menjadi mode "silent," dan nilai 2 akan mengurangi keluaran selama pelatihan. callbacks=[monitor_val_acc]: Ini mengacu pada callback yang telah Anda definisikan sebelumnya, yaitu monitor_val_acc yang menggunakan teknik penghentian dini. Callback ini akan memantau akurasi pada data pelatihan dan menghentikan pelatihan jika akurasi tidak meningkat dalam jumlah epoch yang ditentukan.

```python
history = model.fit(
    x = x_scaled_train, 
    y = y_train,
    validation_data=(x_scaled_test,y_test),
    epochs=500,
    verbose=1,
    callbacks=[monitor_val_acc]
)
```

model.evaluate(x_scaled_test, y_test) digunakan untuk mengevaluasi kinerja model jaringan saraf tiruan (neural network) pada data uji.x_scaled_test: Parameter pertama adalah data input uji (fitur) yang telah diubah skala, sehingga sesuai dengan data pelatihan. Model akan menggunakan data ini untuk membuat prediksi.y_test: Parameter kedua adalah label atau target yang sesuai dengan data uji. 

```python
model.evaluate(x_scaled_test,y_test)
```

selanjutnya adalah pembuatan diagram plot untuk melihat perkembangan grafik akurasi dan loss pada tada yang telah kita training

```python
def make_plot(train, validation, title):
    graph = plt.plot(history.history[train])
    graph = plt.plot(history.history[validation])
    graph = plt.title(title)
    graph = plt.legend(['training','accuracy'])
    graph = plt.show()
    return graph
```

```python
make_plot('loss','val_loss', 'perubahan loss pada tiap epoch')
make_plot('accuracy', 'val_accuracy', 'perubaha akurasi pada tiap epoch')
```
