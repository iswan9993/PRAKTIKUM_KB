    c) seblum melakukan normalisasi kita terlebuh dahulu melakukan pendefinisian fungsi 'scaler = MinMaxScaler()'. dimana normalisasi ini dilakukan pada data yang telah di salin terlebih dahulu.dan pada fungsi 'print' memanggil 'scaler_data' yang menampung perhitungan normalisasi data yang telah di salin sebelumnya begitu juga halnya standar deviasinya. dan perlu diingat bahwa Normalisasi adalah teknik perubahan skala pada data yang memiliki dimensi berbeda-beda menjadi [0, 1].

```python
data = np.asarray([[100, 0.001],
                    [8, 0.05],
                    [50, 0.005],
                    [88, 0.07],])
data_copy = data.copy()

print("Nilai Data Sebelum Normalisasi:")
print(data_copy)
print("Nilai standar deviasi sebelum Normaslisasi:", np.std(data_copy))

scaler = MinMaxScaler()
Scaler_data = scaler.fit_transform(data_copy)

print("\nNilai Data setelah Normalisasi:")
print(Scaler_data)
print("Nilai standar deviasi setelah Normalisasi:", np.std(Scaler_data))
```

    18. untuk merubah nilai null kedalam median,mean dan modus.pada kolom temperatur_celcius saya menggunaka fungsi mean()untuk menghitung rata-rata nilai dalam satu lokasi/modus pada data lokasi yang sering muncul dalam satu waktu dan menggunakan median() atau nilai tengah dalam humidity berdasarkan data yang sering muncul pada location_name  dan setelah di cetak hasilnya data yang sering muncul pada location_name adalah udaipur dengan rata-rata nilai temperature_celciusnya adalah 26.03 dan nilai tengah pada humidity adalah 70.50

```python
df = pd.read_csv('IndianWeatherRepository.csv')

most_frequent_location = df['location_name'].mode()[0]
rata2_suhu = df[df['location_name'] == most_frequent_location]['temperature_celsius'].mean()


rata2_humidity = df[df['location_name'] == most_frequent_location]['humidity'].median()


print(f'Mean temperature_celsius for most frequent location ({most_frequent_location}): {rata2_suhu:.2f}')
print(f'Median humidity for most frequent location ({most_frequent_location}): {rata2_humidity:.2f}')


```

    26.untuk melakukan one-hot-encoding saya menambhakn atribut baru ke dalam dataframe saya dengan nama waktu_hari dan meanbahkan nilai siang dan malam pada labelnya. untuk melakukannya saya menggunakan 'waktu_hari_values = np.random.choice(['siang', 'malam'], size=len(df), p=[0.5, 0.5])' dalam hal ini saya menmbahkan 50% pada siang dan 50% pada malam untuk nilai dalam labelnya.dan memasukkannya kedalam dataframe.

```python
import pandas as pd
import numpy as np


df = pd.read_csv('IndianWeatherRepository.csv')


waktu_hari_values = np.random.choice(['siang', 'malam'], size=len(df), p=[0.5, 0.5])

df['waktu_hari'] = waktu_hari_values


print("DataFrame setelah menambahkan atribut 'waktu_hari':")
print(df.head())


df.to_csv('IndianWeatherRepository.csv', index=False)
```

    27. pertama kita mengimport terlebih dahulu atribut yang diperlukan untuk melakukan one-hot=incoding yaitu 'from sklearn.preprocessing import OneHotEncoder'.dimana pada dataframe kita menginisialisasi 'OneHotEncoder = OneHotEncoder(sparse=False)' terlebih dahulu dan memilih atribut mana yang akan kita one-hot dan disini saya memilih data yang sudah saya buat sebelumnya yaitu waktu_hari,dan menyimpan data kedalam variable SexEnc

```python
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv('IndianWeatherRepository.csv')
OneHotEncoder = OneHotEncoder(sparse=False)
SexEnc = OneHotEncoder.fit_transform(df[["waktu_hari"]])
SexEnc = pd.DataFrame(SexEnc)
df["siang"] = SexEnc[0]
df["malam"] = SexEnc[1]
df.head()
```
